*testPrime() que verifique si un nÃºmero es primo*



```python


m = int(2)
band = "V"

n = int(input("De un nÃºmero:"))
while((band == "V") and (m < n)):
        if((n % m) == 0):
                band = "F"
        else:
                m = m + 1
if(band == "V"):
        print("El nÃºmero es primo")
else:
        print("El nÃºmero no es primo")
```
