***Funciones listas y tuplas***
*Funciones*


*¿Por qué son ventajosas las funciones para tener en sus programas?*

Porque son bloques de sentencias que ejecutan una operación deseada y así se pueden ejecutar varias operaciones.

*¿Cuándo se ejecuta el código en una función: cuando se define la función o cuando se llama a la función?*

La función es lo primero que se define pues si no se realiza no se ejecutará nada, no sólo se deberá definir la función, se le deben asignar parámetros y orden a los mismos sobre los cuales trabajará.

*¿Qué enunciado crea una función?*

Se utiliza la palabra **def** seguida del comando de la función.

*¿Cuál es la diferencia entre una función y una llamada de función?*

**Función**: se le denomina a la secuencia de sentencias que ejecutan alguna operación deseada y tienen un nombre.
**Llamada de función**: contienen el nombre de la función a ejecutar seguida por la
lista de valores, para que la función se ejecute es necesario una llamada de función.

*¿Cuántos ámbitos globales hay en un programa de Python? ¿Cuántos ámbitos locales?*

El ámbito de una *variable global* son todas las funciones que componen el programa.
El ámbito de una *variable local* se restringe a la función que la ha declarado.

*¿Qué sucede con las variables en un ámbito local cuando se devuelve la llamada a la función?*

Cualquier otra función del programa puede conocer el valor de la variable pues la función original debe transferir el valor de la función.

*¿Qué es un valor de retorno? ¿Puede un valor de retorno ser parte de una expresión?*

Es una valor que se utiliza para forzar una salida inmediata de la función en que se encuentra y también devuelve un valor.
No puede ser parte de una expresión pues para que funcione el valor, éste debe ir seguida de una expresión que posteriormente será evaluada.

*Si una función no tiene una declaración de retorno, ¿cuál es el valor de retorno de una llamada a esa función?. *

Si lo omitimos por defecto devuelve un entero.

*¿Cómo puede forzar una variable en una función para referirse a la variable global?*

Se debe anteponer la palabra **extern** a su declaración.

*¿Cuál es el tipo de datos de None?*

Sirve para inicializar una variable previamente declarada si no sabemos antes qué tipo de dato va a cotener o en dónde ningún valor es válido. **NoneType**

*¿Qué hace la declaración import random?*

Importa una serie de funciones relacionadas con los valores aleatorios.

*Si tuviera una función llamada randint() en un módulo llamado random, ¿cómo lo llamaría después de importar random?*

La función randint() devolverá un entero aleatorio en el intervalo comprendido (incluidos los bordes) entre los dos argumentos enteros que le pases. El entero aleatorio devuelto por randint() es almacenado en una variable llamada *número*.

***Listas y tuplas***

*¿Qué es []?*

Se usan para acceder y modificar los elementos de una lista .

*Cómo asignaría el valor 'hola' como el tercer valor en una lista almacenada? en una variable llamada spam? (Supongamos que el spam contiene [2, 4, 6, 8, 10].) Para las siguientes tres preguntas, digamos que spam contiene la lista [2, 4, 6, 8, 10].*

Para asignar un valor en un lugar especifico de la lista se usara **spam.inster(2,'hola')** donde 2:es la posicion en la que se insertara el elemento y spam es el nombre de la lista alamacenada.

*¿Qué evalúa spam[int ('3' * 2) / 11]?*

Arrojará un numero entero de la operación del elemento 3 multiplicado por dos y dividido entre once de la lista.

*¿Qué evalúa spam[-1]?*

Imprimirá el último elemento de la lista en este caso de spam o sea el 10. 

*¿Qué evalúa el spam[: 2]?*

Imprimirá los valores que se encuentran definidos dentro del rango entre 0-2, el cero se encuentra ímplicito.

*Para las siguientes tres preguntas, digamos que bacon contiene la lista [3.14, 'cat', 11, 'cat', True].

    ¿Qué evalúa bacon.index('cat')?*
    
Arrojará la posición del elemento 'cat' sobre la lista bacon.

*¿Qué evalúa bacon.append(99)?*

Se agregará el valor 99 a la lista bacon.

*¿Qué hace bacon.remove('cat')?*

Hace que se quite el elemento cat de la lista bacon.

*¿Cuáles son los operadores para la concatenación de listas y la replicación de listas?*

Podemos concatenar usando el operador de adición **(+)** o a través del método **extend().**

*¿Cuál es la diferencia entre los métodos de lista append() e insert()?*

**Append()** Agrega un ítem al final de la lista
**Insert()** Inserta un ítem en una posición dada. El primer argumento es el índice del ítem delante del cual se insertará

*¿Cuál es la diferencia entre listas y tuplas?*

Una lista puede ser alterada, no así una tupla.
Una tupla puede ser utilizada como clave en un diccionario, no así una lista.
Una tupla consume menos espacio que una lista

*testPrime() que verifique si un número es primo*



```python


m = int(2)
band = "V"

n = int(input("De un número:"))
while((band == "V") and (m < n)):
        if((n % m) == 0):
                band = "F"
        else:
                m = m + 1
if(band == "V"):
        print("El número es primo")
else:
        print("El número no es primo")
```
